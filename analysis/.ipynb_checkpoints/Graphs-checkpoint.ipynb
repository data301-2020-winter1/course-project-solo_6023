```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scripts import m2

data = pd.read_csv("/Users/Mason/cosc301/course-project-solo_6023/data/processed/all_alpha_20.csv")
```

```python
fig=plt.figure(figsize=(14,8), dpi= 100, facecolor='w', edgecolor='k')
countdata = sns.countplot(data=data, x="Type", palette = "crest")
countdata.set(title="Number of 2020 Vehicles by vehicle type", xlabel="Count of Models", ylabel= "Type of Vehciels")
```

```python
fig=plt.figure(figsize=(14,8), dpi= 100, facecolor='w', edgecolor='k')
countdata = sns.countplot(data=data, x="Fuel", palette = sns.dark_palette("#69d", reverse=True, as_cmap=True))
countdata.set(title="Number of 2020 Vehicles by vehicle type", xlabel="Count of Models", ylabel= "Type of Vehciels")
```

```python
fig=plt.figure(figsize=(14,8), dpi= 100, facecolor='w', edgecolor='k')
countdata = sns.regplot(data=data, x="Displacement", y="Combined MPG", logx=True, scatter_kws={"color": "mediumturquoise"}, line_kws={"color": "darkcyan"})
countdata.set(title="Number of 2020 Vehicles by vehicle type", xlabel= "Displacement (in Liters)")
```

```python
fig=plt.figure(figsize=(14,8), dpi= 100, facecolor='w', edgecolor='k')
countdata = sns.stripplot(data=data, x="Combined MPG", y="Type", orient="h", palette="crest")
countdata.set(title="Number of 2020 Vehicles by vehicle type", xlabel= "Combined MPG")
```
